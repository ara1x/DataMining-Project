## **The possible values for each attribute**

```python
attributes_info=[]
for col in data.columns:
 if data[col].dtype=='object':
  possible_values=list(data[col].unique())
  value_range=",".join(map(str,possible_values[:10]))
  attributes_info.append({
    'Attribute': col,
    'Type': 'Categorical',
    'Possible Values': value_range
})
else:
 min_val = data[col].min()
 max_val = data[col].max()
 attributes_info.append({
   'Attribute': col ,
   'type': 'Numerical',
   'possible Values':f"From {min_val} To {max_val}"
})
result=pd.DataFrame(attributes_info)
display(result)
```
